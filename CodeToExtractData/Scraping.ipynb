```python
from bs4 import BeautifulSoup
from urllib2 import urlopen
import datetime
```

```python
#test stub to extract date
url = 'https://www.saatchiart.com/art/Painting-Under-the-sea/25945/2104174/view'
page = urlopen(url)
page=BeautifulSoup(page,'html5lib')
date = page.find("meta",property='bt:pubDate')
date13=date.get("content")
print(date13)

```

```python
def getDate(file1):
    with open(file1) as file:
        for (link) in file:
            page = urlopen(link)
            page=BeautifulSoup(page,'html5lib')
            date = page.find("meta",property='bt:pubDate')
            date13=date.get("content",None)
            date13=date13.encode('utf8')[0:10]
            temp = datetime.datetime.fromtimestamp(float(date13)).strftime('%m-%d-%Y')
            #date123=datetime.fromtimestamp(int(date13).strftime('%m-%d-%y'))
            print(temp)
file='ItemLinks.txt'
print(getDate(file))
```

```python
def getMedium(file1):
    with open(file1) as file:
        for (link) in file:
            page = urlopen(link)
            page=BeautifulSoup(page,'html5lib')
            medium = page.find("meta",property='bt:mediums')
            medium=medium.get("content",None)
            print(medium)
file='ItemLinks.txt'
print(getMedium(file))
```

```python
def getSubject(file1):
    with open(file1) as file:
        for (link) in file:
            page = urlopen(link)
            page=BeautifulSoup(page,'html5lib')
            medium = page.find("meta",property='bt:subject')
            medium=medium.get("content",None)
            print(medium)
file='ItemLinks.txt'
print(getSubject(file))
```

```python
#test stub to calculate brightness of an image
import math
from PIL import Image
from PIL import ImageStat
im_file="pruthvibw70.jpg"
im = Image.open(im_file)
stat = ImageStat.Stat(im)
r,g,b = stat.mean
print(math.sqrt(0.241*(r**2) + 0.691*(g**2) + 0.068*(b**2)))
```

```python
def getImgBrightness(file1):
    with open(file1) as file:
        for (link) in file:
            page = urlopen(link)
            page=BeautifulSoup(page,'html5lib')
            image_link = page.find("meta",property='og:image')
            image_link=image_link.get("content",None)
            #print(image_link)
            #URL='http://saimg-a.akamaihd.net/saatchi/25945/art/2294025/1369168-7.jpg'
            urllib.request.urlretrieve(image_link, "scrpedimg.jpg")
            im_file="scrpedimg.jpg"
            im = Image.open(im_file)
            stat = ImageStat.Stat(im)
            r,g,b = stat.mean
            print(math.sqrt(0.241*(r**2) + 0.691*(g**2) + 0.068*(b**2)))
file='ItemLinks.txt'
print(getImgBrightness(file))
    
```
